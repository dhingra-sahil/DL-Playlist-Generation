```python
print(f"unique songs played: {len(df['songid'].unique())}")
print(f"unique one-off songs: {len(df[df['times_played']==1]['songid'].unique())}")
print(f"unique two-off songs: {len(df[df['times_played']==2]['songid'].unique())}")
print(f"unique three-off songs: {len(df[df['times_played']==3]['songid'].unique())}")
print(f"unique four-off songs: {len(df[df['times_played']==4]['songid'].unique())}")
print(f"unique five-off songs: {len(df[df['times_played']==5]['songid'].unique())}")
```

```python
# because there are so many 1 or 2 -off songs, these are impossible to predict with the historical knowledge
# so they are converted to songid=0 IE "wildcard"
df.loc[df['times_played'] <= 2, 'songid'] = 0
df.loc[df['times_played'] <= 2, 'times_played'] = 510
print(f"unique one-off songs: {len(df[df['times_played']==1]['songid'].unique())}")
print(f"unique two-off songs: {len(df[df['times_played']==2]['songid'].unique())}")
```
